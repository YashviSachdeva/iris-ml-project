# ðŸŒ¸ Iris Flower Prediction Using Linear Regression

This notebook uses basic linear regression to predict **petal length (cm)** from other flower features.

- Dataset: Iris
- Model: Linear Regression
- Input: Sepal Length, Sepal Width, Petal Width
- Output: Petal Length (Predicted)


```python
#Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#Loading the iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# first few rows of the data
df.head()
```

```python
# predicting 'petal length (cm)'
# will use the other columns as inputs
X = df.drop(columns=['petal length (cm)'])  # input features
y = df['petal length (cm)']    # target output
```

```python
#Spliting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

#80% of data is used for training, 20% for testing
```

```python
#Training the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Making predictions on the test data
y_pred = model.predict(X_test)

#printing some predicted values
y_pred[0:5]  # showing first 5 predictions acc to test data
```

```python
# Step 7: Visualize predicted vs actual values
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Petal Length')
plt.ylabel('Predicted Petal Length')
plt.title('Actual vs Predicted')
plt.grid(True)
plt.show()
```

```python
#  User will give input for 3 features (excluding petal length)
sepal_length = float(input("Sepal length (cm) : "))
sepal_width = float(input("Sepal width (cm) : "))
petal_width = float(input("Petal width (cm) : "))

# 3 values to a list mein convert karte hain
input_data = np.array([[sepal_length, sepal_width, petal_width]])

#  Model is predicting
predicted_petal_length = model.predict(input_data)

#Result print
print("Predicted petal length (cm):", predicted_petal_length[0])
```
